```python
import numpy  as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

np.random.seed(42)


n = 300

x = np.random.randn(n)
y = np.random.randn(n)
z = 0.6 * x + 0.2 * y + np.random.randn(n)
datos_3d = np.column_stack((x,y,z))
```

```python
def pca_desde_cero(X, n_componentes):
  #paso 1
  X_centrado = X - np.mean(X, axis=0)
  X_estandarizado = X_centrado / np.std(X, axis=0)

  #paso 2
  cov_matrix = np.cov(X_estandarizado, rowvar=False)

  #paso 3
  eigenvalores, eigenvectores = np.linalg.eig(cov_matrix)

  indices_ordenados = np.argsort(eigenvalores)[::-1]
  eigenvectores_ordenados = eigenvectores[:, indices_ordenados]

  #paso 4

  componentes_principales = eigenvectores_ordenados[:,:n_componentes]

  X_proyectado = X_estandarizado @ componentes_principales

  return X_proyectado

```

```python
#desde la funcion
datos_pca_nuestro = pca_desde_cero(datos_3d, 2)
```

```python
#desde la libreria

scaler = StandardScaler()
datos_escalados_sklearn = scaler.fit_transform(datos_3d)

pca_sklearn = PCA(n_components=2)
datos_pca_sklearn = pca_sklearn.fit_transform(datos_escalados_sklearn)
```

```python
print(f"Resultados PCA: \n{np.round(datos_pca_nuestro[:5],2)}")
print(f"Resultados PCA: \n{np.round(datos_pca_sklearn[:5],2)}")
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))

ax1.scatter(datos_pca_nuestro[:, 0], datos_pca_nuestro[:,1], alpha = 0.7)
ax1.set_title("Nuestro PCA")
ax1.grid(True)

ax2.scatter(datos_pca_sklearn[:, 0], datos_pca_sklearn[:,1], alpha = 0.7)
ax2.set_title("PCA de scikit-leanr")
ax2.grid(True)

plt.suptitle("Validación de nuestro PCA vs Sklearn")
plt.show()
```
