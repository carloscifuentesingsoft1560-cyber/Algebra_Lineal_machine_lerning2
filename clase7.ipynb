```python
import numpy as np

i = np.array([1,0])
j = np.array([0,1])

print(f"Producto pnto entre y j: {i@j}")
```

```python
norma_i = np.linalg.norm(i)
norma_j = np.linalg.norm(j)

print(f"Norma de i es: {norma_i}")
print(f"Norma de j es: {norma_j}")
```

```python
v = np.array([5,3])

v_reconstruido = 5 * i + 3 * j

print(f"¿La base estanda puede contruir este vector V? -> {np.array_equal(v, v_reconstruido)}")
```

```python
v1 = np.array([7,-12])
v2 = np.array([2, 33])
v3 = np.array([3 ,3])

produto_punto_v1_v2 = v1@v2

print(f"Eltroducto punto entre v1 y v2 es: {produto_punto_v1_v2}")
```

```python
produto_punto_v1_v3 = v1 @ v3
 
print(f"Eltroducto punto entre v1 y v3 es: {produto_punto_v1_v3}")
```

```python
norma_v1 = np.linalg.norm(v1)
norma_v2 = np.linalg.norm(v2)
print(f"Norma v1: {norma_v1:.2f}")
print(f"Norma v2: {norma_v2:.2f}")
```

```python
v1_normalizado = v1 / norma_v1
v2_normalizado = v2 / norma_v2

print(f"Vector v1 normalizado: {v1_normalizado}")
print(f"Vector v2 normalizado: {v2_normalizado}")
```

```python
producto_punto_v1_v2_normalizado = v1_normalizado @ v2_normalizado
norma_v1_normalizado = np.linalg.norm(v1_normalizado)
norma_v2_normalizado = np.linalg.norm(v2_normalizado)
```

```python
print(f"producto punto entre v1 y v2 normalizado: {producto_punto_v1_v2_normalizado}")
print(f"norma v1: {norma_v1_normalizado:.2f}")
print(f"norma v2: {norma_v2_normalizado:.2f}")
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,10))
plt.quiver(0,0, v1_normalizado[0], v1_normalizado[1], angles='xy', scale_units = 'xy', scale =1, label='v1 normalizado', color='purple' )
plt.quiver(0,0, v2_normalizado[0], v2_normalizado[1], angles='xy', scale_units = 'xy', scale =1, label='v2 normalizado', color='blue' )
plt.xlim(-5,5);plt.ylim(-5,5)
plt.title('Visualización de vectores ortonormales')
plt.grid(True)
plt.legend()
plt.show()
```
