```python
import numpy as np

I = np.eye(3)

print(f" {I}")
```

```python
A = np.array([
    [1,2,3],
    [4,5,6]

])

print(A @ I)
```

```python
M = np.array([
    [3,1],
    [2,2]

])
```

```python
determinate = np.linalg.det(M)

print(f"Determinate de la matriz M: {determinate:.2f}")
```

```python
M_inversa = np.linalg.inv(M)
print(f"inversa de M {M_inversa}")
```

```python
identidad = M @ M_inversa

print(np.round(identidad,2))
```
