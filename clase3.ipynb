```python
import numpy as np
#precio, calificacion
producto_a = np.array([20,4.5])
print(f"Vector producto{producto_a}")
```

```python
print(f"Numero de ejes o dimensiones del producto:{producto_a.ndim}")
print(f"Forma del producto:{producto_a.shape}")
print(f"Número total de elementos del producto:{producto_a.size}")
```

```python
import matplotlib.pyplot as plt


plt.figure(figsize=(6,6))
plt.quiver(0,0, producto_a[0], producto_a[1], angles='xy', scale_units='xy', scale=1, color='purple')
plt.title('Producto A')
plt.xlabel('Precio')
plt.ylabel('Calificación')
plt.xlim(0, 25)
plt.ylim(0, 10)
plt.grid(True, alpha=0.3)
plt.show()
```

```python
pixel_tensor = np.array([
    [
        [255,0,0],
        [0,255,0],
        [0,0,255]
    ],
    [
        [255,125,100],
        [255,255,40],
        [115,0,255]
    ]
])
print(pixel_tensor.ndim)
print(pixel_tensor.shape)
print(pixel_tensor.size)
```

```python
dias = ['Lunes', 'Martes', 'Miércoles', 'Jueves', 'Viernes', 'Sábado', 'Domingo']
categorias = ['Comida', 'Transporte', 'Ocio', 'Supermercado']
gastos = [[12, 5, 0, 0], 
          [8, 5, 15, 0], 
          [10, 5, 0, 30], 
          [15, 5, 10, 0], 
          [20, 10, 25, 0], 
          [25, 12, 40, 50], 
          [18, 0, 20, 0]]
fig, ax = plt.subplots(figsize=(10, 6))

bottom = np.zeros(len(dias))
for i, cat in enumerate(categorias):
    valores = [dia[i] for dia in gastos]
    ax.bar(dias, valores, bottom=bottom, label=cat)
    bottom += valores

ax.set_title('Mis Gastos Semanales', fontsize=16)
ax.set_ylabel('Monto ($)')
ax.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
