```python
import numpy as np
#filas estudiantes, columnas caracteristicas: hoas estudio, horas de sueño

X = np.array([
    [10, 8],
    [5,7],
    [12,4]
])

w_fila =np.array([[0.3,0.7]])

print(X.shape)
print(w_fila.shape)
```

```python
w_columna = w_fila.T

print(f"w_fila -> {w_fila}")
print(f"w_columna -> {w_columna}")

print(w_fila.shape)
print(w_columna.shape)

print(w_columna)
```

```python
predicciones = X @ w_columna

print(f"prediciones del rendimiento de un estudiante: {predicciones}, formade predicciones {predicciones.shape}")
```
