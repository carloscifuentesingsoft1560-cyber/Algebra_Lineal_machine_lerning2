```python
import gensim.downloader as api

print("Descargando modelo de vectores de palabras")
word_vectors =api.load("glove-wiki-gigaword-50")
print("Modelo cargado")
```

```python
rey = word_vectors['king']
hombre = word_vectors['man']
reina =word_vectors['queen']
gato = word_vectors['cat']
```

```python
def similitud_coseno(v1,v2):
    return v1 @ v2 / (np.linalg.norm(v1)* np.linalg.norm(v2)) 
```

```python
sim_rey_hombre = similitud_coseno(rey, hombre)
sim_rey_reina = similitud_coseno(rey, reina)
sim_rey_gato = similitud_coseno(rey, gato)
```

```python
print(f"similitudo rey y hombre: {sim_rey_hombre:.2f}")
print(f"similitudo rey y reina: {sim_rey_reina:.2f}")
print(f"similitudo rey y gato: {sim_rey_gato:.2f}")
```

```python
# tecnologia, arte, deporte
# para mi amigo y para mi

yo_tas_pelotas = np.array([10,2,9])
amigo_tas_pelotas = np.array([8,5,3])
family_tas_pelotas = np.array([9,1,10])


sim_yo_y_tu = similitud_coseno(yo_tas_pelotas,amigo_tas_pelotas)

print(f"La similitud entre los 2 es de: {sim_yo_y_tu:.2f}")
```

```python
def similitud_coseno_3_vectores(v1,v2,v3):
    def cos_sim(a, b):
        return np.dot(a, b) / (np.linalg.norm(a) * np.linalg.norm(b))
    
    # Calculamos las similitudes por pares
    sim12 = cos_sim(v1, v2)
    sim23 = cos_sim(v2, v3)
    sim13 = cos_sim(v1, v3)
    
    # Retornamos el promedio de las tres
    return (sim12 + sim23 + sim13) / 3
```

```python
sim_yo_fa_ami = similitud_coseno_3_vectores(yo_tas_pelotas,amigo_tas_pelotas,family_tas_pelotas)
print(f"La similitud entre los 3 yo, amigo y mi familia es de : {sim_yo_fa_ami:.2f}")
```
