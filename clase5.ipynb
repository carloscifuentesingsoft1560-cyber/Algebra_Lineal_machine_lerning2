```python
import numpy as np
perfil_ana = np.array([5, 1])
perfil_juan = np.array([1, 5])

norma_ana = np.linalg.norm(perfil_ana)
norma_juan = np.linalg.norm(perfil_juan)

print(f"norma de ana:{norma_ana:.2f}")
print(f"norma de juan:{norma_juan:.2f}")
```

```python
vector = np.array([3,-4,0])
norma_l1 = np.linalg.norm(vector, ord=1)

print(f"La norma L1 del vector es: {norma_l1}")
```

```python
norma_l0 = np.count_nonzero(vector)

print(f"La norma L0 del vector es: {norma_l0}")
```

```python
norma_lif = np.linalg.norm(vector, ord=np.inf)
print(f"La norma L-infinito es:{norma_lif}")
```

```python
casa1 = np.array([120,3])
casa2 = np.array([100,4])

norm_l2_casa1 = np.linalg.norm(casa1)
norm_l2_casa2 = np.linalg.norm(casa2)

print(f"La norma L-2 de la casa 1 es: {norm_l2_casa1:.2f}")
print(f"La norma de L-2 la casa 2 es: {norm_l2_casa2:.2f}")
```
