```python

# crear un modelo de regresion lienal para predecir el precio de una vivienda
# 4 casas con 2 caracteristicas, metros cuadrdos y numero de habitaciones
# OBJETIVO: encontar la linea que mejor se ajusta nos permita predecir el precio

import numpy as np
import matplotlib.pyplot as plt

X = np.array([
    #m^2 y No habitaciones
    #casa1 
    [80, 3],
    #casa2
    [120, 3],
    #casa3
    [100,2],
    #casa4
    [150,4],
    #casa5
    [130,3]
])
#Valor de las casas (1,2,3,4,5) en dolares
Y = np.array([310,390,325,530,400])

```

```python
plt.figure(figsize=(8,6))
plt.scatter(X[:,0], Y, color ='blue', label ='Datos reales de las casas')
plt.xlabel('Metros cuadrados')
plt.ylabel('Precio')
plt.title('¿Po0demos ajustar una linea perfecta en estos datos?')
plt.grid(True)
plt.legend()
plt.show()
```

```python
X_bias = np.c_[np.ones((5,1)), X]
print(X_bias)
```

```python
A = X_bias.T @ X_bias
b = X_bias.T @ Y

print(f"Matiz  A:{A}")
print(f"Vecotor  b:{b}")
```

```python
print(np.linalg.norm(A))
```

```python
A_inversa =np.linalg.inv(A)
print(np.round(A_inversa,2))
```

```python
theta_con_inversa = A_inversa @ b
print(f"Pesos del modelo (theta): {np.round(theta_con_inversa,2)}")
```

```python
theta_con_solve = np.linalg.solve(A,b)
print(f"Pesos del modelo theta con solve:{np.round(theta_con_solve,2)}")
```

```python
theta = theta_con_solve
predicciones = X_bias @ theta

print(f"Precios realies:{Y}")
print(f"Precios predichos:{np.round(predicciones)}")
```

```python
plt.figure(figsize=(8,6))
plt.scatter(X[:, 0],Y, color='blue', label ='Datos reales de las casas')
X_linea = np.array([[50],[160]])
X_linea_bias = np.c_[np.ones((2,1)),X_linea, np.array([[2],[4]])]
Y_linea = X_linea_bias @ theta
print(Y_linea)
plt.plot(X_linea, Y_linea, "r-", label = 'Linea de regresión lineal (nuestro modelo)')
plt.xlabel("Metros cuadrados")
plt.ylabel("Precio en miles")
plt.title("Nuestro modelo de regresión lineal")
plt.grid(True)
plt.legend()
plt.show()
```
