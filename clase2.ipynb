```python
x = np.array([8,9,6,3,2,5])
y = np.array([0,2,5,8,9,6])

v1 = np.array([1,9])
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(6,6))
plt.scatter(x,y, label = 'Puntos de datos')
plt.quiver(0,0, v1[0], v1[1], angles='xy', scale_units='xy', scale=1, color='red', label ='Vector 1')
plt.title('Mi primera visualización')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```
