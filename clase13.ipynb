```python
import numpy as np
import matplotlib.pyplot as plt


X = np.array([

    [80,3],
    [120,3],
    [100,2],
    [150,4]

])

Y = np.array([310,390,325,530])

X_bias = np.c_[np.ones((4,1)),X]

A_saludable = X_bias.T @ X_bias
b = X_bias.T @ Y
```

```python
pies2 = X[:,0] * 12.893

X_enfermo = np.c_[X, pies2]

X_enfermo_bias = np.c_[np.ones((4,1)), X_enfermo]

print(X_enfermo_bias)
```

```python
A_enfermo = X_enfermo_bias.T @ X_enfermo_bias
b = X_enfermo_bias.T @ Y
```

```python
determinante = np.linalg.det(A_enfermo)
print(determinante)
```

```python
rango = np.linalg.matrix_rank(A_enfermo)
print(rango)
```

```python
theta1 = np.linalg.inv(A_enfermo) @ b
print(theta1)
```

```python
theta2 = np.linalg.solve(A_enfermo, b)
print(theta2)
```
